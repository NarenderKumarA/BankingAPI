Filter the DataFrame `df` to include only rows where the 'Question' column is either 'Percent of adults aged 18 years and older who have obesity' or 'Percent of adults who engage in no leisure-time physical activity'. From this filtered data, select the 'Question' and 'Data_Value' columns, drop any rows with missing values in 'Data_Value', and then pivot the data to create two new columns, 'Obesity' and 'Physical Inactivity', corresponding to the respective 'Question' values for analysis in RQ1.

## Filter and Prepare Data for RQ1

### Subtask:
Filter the dataset to include only rows relevant to 'Percent of adults aged 18 years and older who have obesity' and 'Percent of adults who engage in no leisure-time physical activity'. Clean this subset by handling any missing 'Data_Value' entries. Pivot or reshape the data to create distinct columns for the independent and dependent variables for RQ1.


**Reasoning**:
The subtask requires filtering the main DataFrame `df` to isolate data relevant to RQ1, which involves specific 'Question' categories. After filtering, it's necessary to select relevant columns, handle missing 'Data_Value' entries, pivot the data to create distinct columns for the independent and dependent variables, rename these columns for clarity, and finally, drop any rows that still contain missing values after the pivot, ensuring a clean dataset for regression analysis.



```python
import pandas as pd

# 1. Define a list of the two specific 'Question' values relevant to RQ1
rq1_questions = [
    'Percent of adults aged 18 years and older who have obesity',
    'Percent of adults who engage in no leisure-time physical activity'
]

# 2. Filter the DataFrame `df` to include only the rows where the 'Question' column matches one of the values in the list
df_rq1_filtered = df[df['Question'].isin(rq1_questions)].copy()

# 3. From this filtered DataFrame, select the columns 'YearStart', 'LocationAbbr', 'LocationDesc', 'StratificationCategory1', 'Stratification1', 'Question', and 'Data_Value'
df_rq1_selected = df_rq1_filtered[[
    'YearStart',
    'LocationAbbr',
    'LocationDesc',
    'StratificationCategory1',
    'Stratification1',
    'Question',
    'Data_Value'
]].copy()

# 4. Drop any rows from the selected columns that have missing values in the 'Data_Value' column
df_rq1_cleaned = df_rq1_selected.dropna(subset=['Data_Value']).copy()

# 5. Pivot the cleaned data
df_rq1 = df_rq1_cleaned.pivot_table(
    index=['YearStart', 'LocationAbbr', 'LocationDesc', 'StratificationCategory1', 'Stratification1'],
    columns='Question',
    values='Data_Value'
).reset_index()

# 6. Rename the columns
df_rq1 = df_rq1.rename(columns={
    'Percent of adults aged 18 years and older who have obesity': 'Obesity',
    'Percent of adults who engage in no leisure-time physical activity': 'Physical Inactivity'
})

# 7. Drop any rows in df_rq1 that still contain missing values
df_rq1.dropna(inplace=True)

print("Prepared DataFrame for RQ1:")
print(df_rq1.head())
print("\nShape of df_rq1:", df_rq1.shape)
print("\nMissing values in df_rq1:")
print(df_rq1.isnull().sum())
```

## Build and Evaluate Regression Model for RQ1

### Subtask:
Formulate null and alternative hypotheses for Revised RQ1. Build a linear regression model with 'Percent of adults aged 18 years and older who have obesity' as the independent variable and 'Percent of adults who engage in no leisure-time physical activity' as the dependent variable. Evaluate the model's performance by examining coefficients, R-squared, and p-values.


### Hypotheses for Revised Research Question 1 (RQ1):

**Revised Research Question 1 (RQ1):** How well does the percentage of adults who have obesity predict the percentage of adults who engage in no leisure-time physical activity?

*   **Null Hypothesis (H0):** There is no significant linear relationship between the percentage of adults who have obesity and the percentage of adults who engage in no leisure-time physical activity. The coefficient for 'Obesity' in the regression model is zero (1 = 0).

*   **Alternative Hypothesis (H1):** There is a significant linear relationship between the percentage of adults who have obesity and the percentage of adults who engage in no leisure-time physical activity. The coefficient for 'Obesity' in the regression model is not zero (1  0).

**Reasoning**:
Now that the hypotheses are stated, I will build and evaluate the linear regression model for RQ1 as per the instructions. This involves importing necessary libraries, constructing the OLS model with 'Physical Inactivity' as the dependent variable and 'Obesity' as the independent variable from `df_rq1`, and then printing the model summary to evaluate its performance.



```python
import statsmodels.formula.api as smf
import statsmodels.api as sm

# Build the linear regression model for RQ1
# Dependent variable: 'Physical Inactivity'
# Independent variable: 'Obesity'
model_rq1 = smf.ols('Q("Physical Inactivity") ~ Q("Obesity")', data=df_rq1).fit()

# Print the summary of the model
print("\nLinear Regression Model Summary for RQ1:")
print(model_rq1.summary())
```

**Reasoning**:
The next step in the overall task is to visualize the results of the linear regression model. This involves creating a scatter plot of 'Obesity' versus 'Physical Inactivity' and overlaying the regression line from the `model_rq1` to visually represent the relationship, as requested in the overall task description.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot for RQ1
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Obesity', y='Physical Inactivity', data=df_rq1, alpha=0.6)

# Add the regression line to the scatter plot
sns.regplot(x='Obesity', y='Physical Inactivity', data=df_rq1, scatter=False, color='red', line_kws={'label': 'Regression Line'})

# Add title and labels
plt.title('Relationship between Obesity and Physical Inactivity (RQ1)')
plt.xlabel('Percent of adults aged 18 years and older who have obesity')
plt.ylabel('Percent of adults who engage in no leisure-time physical activity')
plt.legend()
plt.grid(True)
plt.show()
```

## Method Discussion

### Linear Regression Explained

Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables by fitting a linear equation to observed data. In its simplest form (simple linear regression), it seeks to establish a straight line that best describes how a change in the independent variable is associated with a change in the dependent variable. The mathematical representation of a simple linear regression model is typically given by:

\[ Y = \beta_0 + \beta_1 X + \epsilon \]

Where:
*   `Y` represents the dependent variable (the outcome being predicted).
*   `X` represents the independent variable (the predictor).
*   `\beta_0` is the y-intercept, representing the expected value of `Y` when `X` is zero.
*   `\beta_1` is the coefficient for `X`, representing the change in `Y` for each one-unit change in `X`.
*   `\epsilon` is the error term, accounting for the variability in `Y` that cannot be explained by `X`.

The goal of linear regression is to estimate the unknown parameters (\(\beta_0\) and \(\beta_1\)) using observed data, typically through the method of least squares, which minimizes the sum of the squared differences between the observed and predicted values.

### Suitability for RQ1 and RQ2

Linear regression was chosen as the appropriate analytical tool for both Revised Research Question 1 (RQ1) and Revised Research Question 2 (RQ2) due to the nature of the variables involved and the objectives of the analysis:

**Revised Research Question 1 (RQ1):** "How well does the percentage of adults who have obesity predict the percentage of adults who engage in no leisure-time physical activity?"

*   **Dependent Variable (Y):** 'Percent of adults who engage in no leisure-time physical activity' (Continuous).
*   **Independent Variable (X):** 'Percent of adults aged 18 years and older who have obesity' (Continuous).

**Revised Research Question 2 (RQ2):** "How well does the percentage of adults who achieve at least 150 minutes a week of moderate-intensity aerobic physical activity predict the percentage of adults who report consuming fruit less than one time daily?"

*   **Dependent Variable (Y):** 'Percent of adults who report consuming fruit less than one time daily' (Continuous).
*   **Independent Variable (X):** 'Percent of adults who achieve at least 150 minutes a week of moderate-intensity aerobic physical activity...' (Continuous).

In both research questions, we are investigating the predictive relationship between two continuous variables. The objective is to quantify how a change in one variable is linearly associated with a change in the other. Linear regression is ideally suited for this purpose, as it allows us to:

1.  **Quantify the Relationship:** The regression coefficient (\(\beta_1\)) provides a direct measure of the strength and direction of the linear association. For RQ1, a positive coefficient would indicate that higher obesity rates are associated with higher physical inactivity. For RQ2, a negative coefficient would suggest that higher physical activity levels are associated with lower rates of infrequent fruit consumption.
2.  **Assess Predictive Power:** The R-squared value indicates the proportion of the variance in the dependent variable that can be predicted from the independent variable, offering insights into the model's overall explanatory power.
3.  **Test for Statistical Significance:** The p-value associated with the regression coefficients allows us to determine if the observed relationships are statistically significant, helping us to accept or reject the null hypotheses formulated for each question.

Given the continuous nature of both the predictor and outcome variables and the goal of understanding and quantifying their linear association, linear regression provides a straightforward and interpretable framework for addressing these research questions.
